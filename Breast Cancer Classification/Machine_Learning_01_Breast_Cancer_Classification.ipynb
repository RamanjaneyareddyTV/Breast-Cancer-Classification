# STEP Nº4: MODEL TRAINING (FINDING A SOLUTION)

```python
# drop the target column from the dataset to create the input of the model
X = df_cancer.drop(['target'], axis = 1)
# preview the dataset
X
```

```python
# create the expected output
y = df_cancer['target']
# preview data
y
```

```python
# import model for training the model
from sklearn.model_selection import train_test_split
```

```python
# split data in training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
# memo: To see the docstring help insert the ? up the end of the fuction and the run
# Ex-- > train_test_split?
```

```python
# preview data
X_train
```

```python
# import supor vector machine model from sklearn
from sklearn.svm import SVC
```

```python
# import classification_report and confusion
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#define an instant of SVC
svn_model= SVC()

# Train the model
svn_model.fit(X_train, y_train)
```

# STEP Nº 5: EVALUATING  THE MODEL

```python
# predict the output of our model according using the testing dataset
y_predict = svn_model.predict(X_test)
```

```python
y_predict
```

```python
# calculate the confusion matrix in order to evaluate the results
cm = confusion_matrix(y_test, y_predict)
```

```python
#show comparizon using Heat Map
sns.heatmap(cm, annot = True)
# first Iteration: The result on the training is terrible. We need to improve the model 
```

```python
# Classification results
print(classification_report(y_test, y_predict))
```

# STEP Nº6: IMPROVING THE MODEL (only normalization)





```python
# calculate the min value
min_train = X_train.min()
```

```python
# calculate the range of all the set
range_train = (X_train - min_train).max()
# use the min and the range to normalize the data
X_train_scaled = (X_train - min_train)/range_train
```

```python
# Figure 1: Train data not scaled
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
# Figure 2: Train data scaled
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
# normalize testing dataset-----
# calculate the min value
min_test = X_test.min()
# calculate the range of all the set
range_test = (X_test - min_test).max()
# use the min and the range to normalize the data
X_test_scaled = (X_test - min_test)/range_test
```

```python
# Figure 3: Test data scaled
sns.scatterplot(x = X_test_scaled['mean area'], y = X_test_scaled['mean smoothness'], hue = y_test)
```

```python
# Train the model again
svn_model.fit(X_train_scaled, y_train)
```

```python
# predict the output of our model according using the scaled testing dataset
y_predict = svn_model.predict(X_test_scaled)
```

```python
# calculate the confusion matrix in order to evaluate the results
cm = confusion_matrix(y_test, y_predict)
```

```python
#Compare again using Heat Map
sns.heatmap(cm, annot = True)
# first Iteration: The result on the training is terrible. We need to improve the model 
```

```python
# Classification results
print(classification_report(y_test, y_predict))
```

# STEP Nº7: IMPROVING THE MODEL (Gamma and C value)

```python
'''
C and Gamma are the parameters for a nonlinear support vector machine (SVM) with a Gaussian radial basis function kernel (rbf).
A standard SVM seeks to find a margin that separates all positive and negative examples. However, this can lead to poorly fit models if any examples are mislabeled or extremely unusual.
to find the est parameters that fit our model we use GridSearchCV and then used on the predict functions.
'''


# set a list of C and Gamma values
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001], 'kernel': ['rbf']}
```

```python
# import method to calculate the best C and Gamma value
from sklearn.model_selection import GridSearchCV
```

```python
# set the calculation for the best parameters
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
```

```python
# train trying to fit the best parameters 
grid.fit(X_train_scaled, y_train)
```

```python
# shows our best values
grid.best_params_
```

```python
# predict 
grid_y_predict = grid.predict(X_test_scaled)
```

```python
# calculate the confusion matrix in order to evaluate the results
cm = confusion_matrix(y_test, grid_y_predict)
```

```python
#Compare again using Heat Map
sns.heatmap(cm, annot = True)
```

```python
# Classification results
print(classification_report(y_test, grid_y_predict))
```
